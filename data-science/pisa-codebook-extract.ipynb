```python
import xlrd
import wget
import re
import pandas as pd
```

```python
url = 'https://www.oecd.org/pisa/data/2015database/Codebook_CMB.xlsx'
loc = 'data/Codebook_CMB.xlsx'
```

```python
wb = []

try:
    wb = xlrd.open_workbook(loc)
except e:
    wget.download(url, loc)

sheet = wb.sheet_by_name('Cognitive')
```

```python
values = [sheet.row_values(i) for i in range(sheet.nrows) if sheet.cell_value(i,0) != '']
```

```python
df = pd.DataFrame(values[22:-1],columns=values[0])
df
```

```python
df[['MIN','MAX']] = df['MINMAX'].str.split("-",expand=True)
df[['TYPE']] = df['VARLABEL'].str.extract('.*\((.*)\).*')
df
```
